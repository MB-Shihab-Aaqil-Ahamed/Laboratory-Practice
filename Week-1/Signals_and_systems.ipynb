```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
fig, ax = plt.subplots(figsize=(12,6))

A = 1
omega0 = 1
t = np.arange(-3*np.pi, 3*np.pi + np.pi/100 , np.pi/100)
phi = -np.pi/6
xt1 = A*np.cos(omega0*t + phi)

phi = -np.pi/2
xt2 = A*np.cos(omega0*t + phi)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.set_xlabel('$t/\pi$')
ax.set_ylabel('$x(t)$')
ax.plot(t/np.pi, xt1, label='$A\cos(t-\pi/6)$')
ax.plot(t/np.pi, xt2, label='$A\cos(t-\pi/6)$')
ax.grid(True)
ax.legend(loc='lower right')
plt.show()
```

# DT Sinusoid ($N=12$)
$A=1, \omega_0= 2\pi/N$
1. $\phi = 0$
2. $\phi = -\pi/2$

```python
fig, ax = plt.subplots(figsize=(12,6))
n = np.arange(-48,48 + 1, 1)
A = 1
N = 12
omega0 = 2*np.pi/N
phi = -np.pi/2
xn = A*np.cos(omega0*n + phi)
ax.stem(n, xn, linefmt = 'b', markerfmt = 'bo')

ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.set_xlabel('$n$')
ax.set_ylabel('$x[n]$')
plt.show()
```

# Exoinential Signal $x(t) = Ce^{at + t_0}$

```python
fig, ax = plt.subplots(1, 2, figsize=(12,6))

t = np.arange(-1, 3, 0.01)

C = 1
t0 = 0
a = 1 #positive
xt1 = C*np.exp(a*t + t0)

a = 1.2
xt2 = C*np.exp(a*t + t0)

C = 1
t0 = 0
a = -1 #negative
xt3 = C*np.exp(a*t + t0)

a = -1.2
xt4 = C*np.exp(a*t + t0)

ax[0].plot(t, xt1, label='$e^{t}$')
ax[0].plot(t, xt2, label='$e^{1.2t}$')
ax[0].grid(True)
ax[0].legend(loc='lower right')

ax[1].plot(t, xt3, label='$e^{t}$')
ax[1].plot(t, xt4, label='$e^{1.2t}$')
ax[1].grid(True)
ax[1].legend(loc='lower right')
plt.show()
```
